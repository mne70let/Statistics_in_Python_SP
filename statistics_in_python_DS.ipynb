# Закон Больших Чисел (Low of Large numbers)

* Если случайные величины независимы
$$ X_1, X_2, X_3 \cdots X_n \sim iid $$
* Если существует мат ожидание $ \mathbb{E}(X) = \mu $
* Если дисперсия конечна $ \mathbb{E}(X^2) < \infty $

```python
np.random.seed(42)
x = np.random.randint(1, 7, 1000) #бросок кубика
x_cum = np.cumsum(x) #накопленная сумма 
n = np.arange(1, len(x)+1) #return evenly spaced values within a given interval.
X = x_cum / n #среднее на n-ом броске
X[:10]
```

```python
plt.plot(n, X)
plt.grid(True)
plt.axhline(3.5, c='r');
plt.axis([0, len(x)+1, 2, 6])
```

Confidence Interval = x  +/-  t*(s/√n)

where:

    x: sample mean
    t: t-value that corresponds to the confidence level
    s: sample standard deviation
    n: sample size


```python
smpl = np.random.choice(population, 10)
```

Доверительный интервал:

```python
sts.t.interval(alpha=0.95, df=len(smpl)-1, loc=np.mean(smpl), scale=sts.sem(smpl) )
```

```python
min_n = 1
max_n = 3

mins = np.zeros(max_n-min_n+1)
maxs = np.ones(max_n-min_n+1)
ns = np.arange(min_n, max_n+1)

plt.plot(ns, mins, ns, maxs);
```

```python
min_n = 10
max_n = 100

mins = np.zeros(max_n-min_n+1)
maxs = np.ones(max_n-min_n+1)
ns = np.arange(min_n, max_n+1)

for n in range (min_n,max_n+1):
    smpl = np.random.choice(population, n, replace=True)
    conf_int = sts.t.interval(alpha=0.95, df=len(smpl)-1, loc=np.mean(smpl), scale=sts.sem(smpl) )
    mins[n-min_n]=conf_int[0]
    maxs[n-min_n]=conf_int[1]

plt.plot(ns, mins, ns, maxs); 
plt.axhline(np.mean(population), c='r');
```

## __Вселенная 25 (мыши)__

Мыши-самки:

```python
pop1 = sts.norm(loc=2, scale=1).rvs(size=100000)
print(np.mean(pop1))
print(np.std(pop1))
```

Мыши-самцы:

```python
pop2 = sts.norm(loc=5, scale=2).rvs(size=100000)
print(np.mean(pop2))
print(np.std(pop2))
```

```python
def conf_int_norm(x, alpha=0.05):
  return sts.norm.interval(
      1-alpha, x.mean(),
      x.std(ddof=1)/np.sqrt(x.shape[0])
      )
```

```python
np.random.seed(42)
n=5
x_1 = np.random.choice(pop1, size=n)
x_2 = np.random.choice(pop2, size=n)
x1_ci = conf_int_norm(x_1)
x2_ci = conf_int_norm(x_2)

plt.figure(figsize=(15,10))
plt.hist(pop1, bins=50, alpha=.5, density=True)
plt.axvline(x1_ci[0], color='b', ls="--")
plt.axvline(x1_ci[1], color='b', ls="--")
plt.hist(pop2, bins=50, alpha=.5, density=True)

plt.axvline(x2_ci[0], color='r', ls="--")
plt.axvline(x2_ci[1], color='r', ls="--")
plt.show()
```
